전처리용 convert 함수 - Cristiano Ronaldo -> C. Ronaldo

```python
def convert(n):
    div_str = list(n.split(" "))

    if len(div_str) == 1:
        return n
    else:
        return div_str[0][0] + '. ' + div_str[len(div_str) - 1]
        #return div_str[len(div_str) - 1]
```
