```python
def convert(n):
    div_str = list(n.split(" "))

    if len(div_str) == 1:
        return n
    else:
        regex_name_list = re.findall("[a-zA-Z]+", n)
        return regex_name_list[0][0] + '. ' + regex_name_list[len(div_str) - 1]

n = 'David de Gea'
convert(n)
```
