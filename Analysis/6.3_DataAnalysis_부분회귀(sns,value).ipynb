```python
others = list(set(df_p.columns).difference(set(["value", "follower"])))
p, resids = sm.graphics.plot_partregress(
    "value", "follower", others, data=df_p, obs_labels=False, ret_coords=True
)
plt.show()
```

```python
dfX0 = df_p.drop('value', axis=1)
dfX = sm.add_constant(dfX0)
dfy = df_p['value']
df = pd.concat([dfX, dfy], axis=1)

model = sm.OLS(dfy, dfX)
result = model.fit()
```
